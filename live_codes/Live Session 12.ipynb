```python
# color style
x = np.linspace(0, 20, 100)

plt.figure(figsize=(12, 8))
plt.plot(x, np.sin(x), color='green', label='Sin x') # menggunakan nama dari warnanya
plt.plot(x, np.sin(x-1), color='m', label='Sin X-1') # format rgbcmyk
plt.plot(x, np.sin(x-2), color='0.5', label='Sin X-2') # grayscale 0 1
plt.plot(x, np.sin(x-3), color='#FF0037', label='Sin X-3') # RRGGBB , value from 00 to FF
plt.plot(x, np.sin(x-4), color=(0, 1, 0.5), label='Sin X-4') # (R, G, B) 0 - 1
plt.plot(x, np.sin(x-5), color='DarkGreen', label='Sin X-5') # html color names supports
plt.xlabel('X Value')
plt.ylabel('Sin X')
plt.title('Macam macam\nSinx')
plt.legend(loc='lower center', framealpha=0.01, ncol=3, bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# line style
plt.figure(figsize=(12, 8))
plt.plot(x, x, linestyle='solid') # solid
plt.plot(x, x+1, linestyle='--') # dashed
plt.plot(x, x+2, linestyle='-.') # dashdotted
plt.plot(x, x+3, linestyle=':') # dotted
# plt.plot(x, x+8, linestyle=None, marker='x')
plt.xlim([5, 15])
plt.ylim([5, 20])
plt.show()
```

```python
rng = np.random.RandomState(0)
```

```python
list_marker = ['x', 'o', '.', '+', '<', '>', 'd']
plt.figure(figsize=(12, 8))
for marker in list_marker :
    plt.plot(rng.rand(7), rng.rand(7), marker=marker, linestyle='None',
             label="marker= {}".format(marker)
            )
plt.legend()
plt.show()
```
