```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
pd.pivot(titanic, index='sex', columns='class', values='age')
```

```python
df = pd.DataFrame({
    'pclass' : ['first', 'second', 'third', 'fourth'],
    'sex' : ['male', 'female', 'male', 'female'],
    'age' : [12, 20, 15, 28]
})
```

```python
df.head()
```

```python
df_2 = df.pivot(index='sex', columns='pclass', values='age')
```

```python
df_2.reset_index()
```

```python
df_2.reset_index(inplace=True)
```

```python
pd.melt(df_2, id_vars=['sex'], value_vars=['first', 'fourth'])
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['age'].mean()
```

```python
titanic.groupby(['sex', 'survived'])['age'].mean()
```

```python
titanic.head()
```

```python
df_3 = pd.pivot_table(titanic, index='sex', columns=['adult_male', 'alone'], values=['survived', 'age'], aggfunc=['sum'], margins=True)
```
