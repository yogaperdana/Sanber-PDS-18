<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

*Built-in function* adalah fungsi bawaan yang telah tersedia di interpreter Python dan dapat langsung digunakan

Tiga contoh *built-in function* di Python diantaranya:

1.  **sum()**

    Fungsi untuk **menjumlahkan** semua nilai yang ada pada suatu variabel data yang dapat diiterasi (*list* dan *tuple*)

2.  **min()**

    Fungsi untuk mendapatkan **nilai terkecil** pada suatu variabel data yang dapat diiterasi (*list* dan *tuple*)

3.  **max()**

    Fungsi untuk mendapatkan **nilai terbesar** pada suatu variabel data yang dapat diiterasi (*list* dan *tuple*)

```python
# Contoh penggunaan built in function
data = (12, 24, 35, 6, 99, 50)

print(str(data))
print('Jumlah: ' + str(sum(data)))
print('Nilai terkecil: ' + str(min(data)))
print('Nilai Terbesar: ' + str(max(data)))
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

Fungsi (*function*) adalah sebuah blok kode yang membungkus dan membawa tugas/perintah tertentu. Blok tersebut dapat dipanggil dengan menuliskan nama fungsinya.

Metode (*method*) mirip dengan *function* namun lebih terasosiasi pada sebuah *object*. *Method* dipanggil pada suatu *object* dan dapat membuat perubahan pada *object* tersebut. Sebuah *method* mengakses fungsi yang ada di dalam kelas (*class*) yang telah didefinisikan.

Contoh:


```python
class Kelas(object):
    def metodeku(self):
        print('Hai')

def fungsiku():
    print('Halo')

hai = Kelas()
hai.metodeku() #<-- memanggil method dari dalam kelas

fungsiku() #<-- memanggil fungsi
```

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    return sum(inp_list)/len(inp_list)
    
print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(list1, list2):
    return [*list1, *list2]
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
