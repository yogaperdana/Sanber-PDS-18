<h2>Soal 1: Membuat Objek DataFrames</h2>

- Buatlah objek dataframes dari objek dictionary, Bebas!

```python
# Membuat dataframes dari dictionary
import pandas as pd

konfirmasi = {
    'Kota Yogyakarta': [346, 51, 18],
    'Kab. Sleman': [832, 409, 22],
    'Kab. Bantul': [602, 70, 21],
    'Kab. Kulon Progo': [117, 60, 5],
    'Kab. Gunungkidul': [212, 21, 9],
    'Non DIY': [14, 4, 0]
}

dd = {'Wilayah': [], 'Sembuh': [], 'Dirawat': [], 'Meninggal': []}
for key, values in konfirmasi.items():
    dd['Wilayah'].append(key)
    dd['Sembuh'].append(values[0])
    dd['Dirawat'].append(values[1])
    dd['Meninggal'].append(values[2])

df = pd.DataFrame(dd)
df
```

<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

**Jawab**:

*   Perbedaan *Series* dengan *list* dan *dictionary* adalah pada tipe data dan dimensinya. Data *value* pada *series* harus bertipe sama, sedangkan pada *list* dan *value* pada *dictionary* dapat berbeda tipe data. Objek *series* hanya bersifat satu dimensi saja.

*   *Series* adalah objek 1 dimensi sedangkan *dataframe* adalah objek 2 dimensi. Setiap kolom pada *dataframe* merupakan objek dari *series*, dan baris terdiri dari elemen yang ada pada *series*.

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

**Jawab:**

Statistika deskriptif adalah metode-metode yang berkaitan dengan pengumpulan dan penyajian suatu gugus data sehingga memberikan informasi yang berguna **(Walpole, 1993)**. Statistik deskriptif dapat mendeskripsikan data yang sedang dianalisis, tetapi tidak boleh menarik kesimpulan apapun dari data **(Yulianto, 2012)**.

Ukuran-ukurannya dapat digolongkan menjadi dua kelompok, yaitu ukuran nilai tengah dan ukuran deviasi. Ukuran nilai tengah terdiri dari rata-rata (*mean*), median dan modus. Sedangkan ukuran deviasi terdiri dari varians, simpangan baku, koefisien variasi dan nilai jarak (*range*).

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
dfcsv.describe()
```

```python
print('nilai rata-rata dari column price:', dfcsv.price.mean())
print('nilai standar deviasi dari column depth:', dfcsv.depth.std())
print('nilai maksimum dari column carat:', dfcsv.carat.max())
```

**Jawab:**
*   Nilai rata-rata dari *column* `price` adalah **3932.799721913237**
*   Nilai standar deviasi dari *column* `depth` adalah **1.4326213188337733**
*   Nilai maksimum dari *column* `carat` adalah **5.01**
