**Jawab:**

1.   *Machine learning* adalah sebuah metode sistem yang membuat suatu prediksi atau keputusan dari berbagai data masukan dengan mempelajari berbagai modelnya.

2.   *Data feature* (fitur) adalah variabel masukan yang digunakan dalam membuat prediksi.
     *Data target* adalah variabel keluaran yang akan menjadi hasil prediksi.

3.   *Supervised learning* adalah pembelajaran pada mesin dengan menggunakan data yang telah diberi label sehingga identifikasi relasi pada data dapat dilakukan berdasarkan labelnya, sedangkan *unsupervised learning* menggunakan data yang tidak ada labelnya sehingga pengelompokkan data dilakukan dengan mengelompokkan data fitur yang memiliki sifat dan atribut yang sama.

4.   Beberapa jenis *supervised learning*:

     *   Klasifikasi

         Mengkategorikan beberapa variabel ke dalam sebuah set kelas diskrit (kelas yang tidak saling berhubungan)

     *   Regresi

         Mengidentifikasi relasi atau hubungan diantara dua variabel atau lebih untuk memprediksi nilai kontinu.

5.   Hyperparameter adalah variabel yang nilainya sudah didefinisikan di awal dan tidak diubah selama proses *training* serta tidak bergantung pada data, sedangkan parameter nilainya bisa didapatkan secara langsung saat proses *trainin*g model.

6.   Langkah-langkah dalam mengaplikasikan algoritma pada *machine learning*:

     *    Pilih model atau algoritma yang akan digunakan

     *    Definisikan *hyperparameter*

     *    Pisahkan antara data *feature* dan data *target*

     *    Perintahkan model untuk mempelajari data dengan menggunakan metode `.fit()`
     
     *    Applikasikan model terhadap data *test* dengan menggunakan metode `.predict()`

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

- n features : 1
  
Tree Graph :

```python
from sklearn import tree

dtr = tree.DecisionTreeRegressor(max_depth=4, random_state=0, splitter='random')

x_train = x.reshape(-1, 1)
y_train = y
x_test = x_new.reshape(-1, 1)

dtr.fit(x_train, y_train)
y_test = dtr.predict(x_test)
y_predict = dtr.predict(x_train)

fig = plt.subplots(figsize=(12, 8))
plt.scatter(x, y, c='b', label = 'Data Asli')
plt.scatter(x_train, y_predict, c='y', label = 'Prediksi Terhadap Data Training')
plt.plot(x_test, y_test, c='g', label = 'Prediksi Terhadap Data Baru')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1.15), frameon=True)
plt.show()
```

```python
print('feature importance:', dtr.feature_importances_)
```

```python
print('n features:', dtr.n_features_)
```
